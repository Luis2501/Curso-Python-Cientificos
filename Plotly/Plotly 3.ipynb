```python
import numpy as np
import plotly.graph_objects as go 
from plotly.subplots import make_subplots
```

```python
x = np.linspace(-2,2,1001)

fig = make_subplots(rows=1, cols=2, subplot_titles=("Gr√°fica 1", "Gr√°fica 2"))

fig.add_trace(go.Scatter(x = x, y= x**2, name="Funci√≥n cuadratica", mode="lines", line_color="green"), 
              row=1, col=1)

fig.add_trace(go.Bar(x=["Peras", "Manzanas", "Sandias"], y=[1,2,3], name="Gr√°fica de Barras"), row= 1, col=2)

fig.update_layout(title="Subplots")
```

```python
fig.write_html("Prueba.html")
```

```python

# plotly, plotly_white, plotly_dark ggplot2

fig.update_layout(template="ggplot2")
fig.show()
```

```python
fig.update_xaxes(title_text="Eje x", row=1, col=1)
fig.update_xaxes(title_text="Eje x", row=1, col=2)

fig.update_yaxes(title_text="Eje y", row=1, col=1)
fig.update_yaxes(title_text="Eje y", row=1, col=2)

fig.show()
```
